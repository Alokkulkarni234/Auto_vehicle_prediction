## Model Creation

```python
#Step 1 Independent and dependent variable creation
X=df[['normalized_losses','fuel-type','aspiration', 'body-style', 'drive-wheels', 'length', 'width', 'curb-weight', 'num-of-cylinders', 'engine_size', 'fuel-system', 'bore','horsepower', 'city-mpg']] ## X variable contains all inputs
Y=df.price ## Y is output
```

```python
# Train Test Split

from sklearn.model_selection import train_test_split
X_train,X_test,Y_train,Y_test= train_test_split(X,Y,test_size=0.3,random_state=42)
```

```python
# Using Robust Scalar :
from sklearn.preprocessing import RobustScaler
scaler = RobustScaler()

X_train_rf = scaler.fit_transform(X_train)
X_test_rf = scaler.transform(X_test)
```

```python
import xgboost as xgb
from sklearn.model_selection import GridSearchCV
from sklearn.metrics import r2_score

# Define the parameter grid
param_grid = {
    'n_estimators': [50, 100, 200],  # Number of boosting rounds
    'learning_rate': [0.01, 0.05, 0.1],  # Step size shrinkage 
    'max_depth': [3, 4, 5],  # Maximum depth of a tree
    'colsample_bytree': [0.6, 0.8, 1.0],  # Subsample ratio of columns when constructing each tree
    'subsample': [0.6, 0.8, 1.0],         # Subsample ratio of the training instance
    'gamma': [0, 0.1, 0.2],               # Minimum loss reduction required to make a further partition on a leaf node
    'reg_alpha': [0, 0.1, 1],             # L1 regularization term on weights
    'reg_lambda': [0, 0.1, 1],            # L2 regularization term on weights
}

# Create XGBoost Regressor
xgb_model = xgb.XGBRegressor(objective='reg:squarederror', eval_metric='rmse', random_state=42) #Add eval_metric and objective

# Instantiate GridSearchCV
grid_search = GridSearchCV(estimator=xgb_model, param_grid=param_grid, cv=3, scoring='r2', verbose=1, n_jobs=-1) #n_jobs to use all processors

# Fit the grid search
grid_search.fit(X_train_rf, Y_train)

# Print the best parameters and score
print("Best parameters found: ", grid_search.best_params_)
print("Best R-squared score: ", grid_search.best_score_)

# Get the best model
best_xgb_model = grid_search.best_estimator_

# Make predictions on the test set
Y_pred = best_xgb_model.predict(X_test_rf)

# Evaluate the model
r2 = r2_score(Y_test, Y_pred)
print("R-squared on the test set:", r2)
```

## Model Evaluation

```python
from sklearn.metrics import mean_squared_error

# test set RMSE
test_rmse = mean_squared_error(Y_test, Y_pred) ** (1 / 2)
 
# Print rmse
print('Root mean Square error: {:.2f}'.format(test_rmse))
```

```python
#Training Score

# Predict on train set
Y_Train_pred = best_xgb_model.predict(X_train_rf)

# train set RMSE
train_rmse = mean_squared_error(Y_train, Y_Train_pred) ** (1 / 2)
 
# Print rmse
print('Root mean Square error: {:.2f}'.format(train_rmse))
```
